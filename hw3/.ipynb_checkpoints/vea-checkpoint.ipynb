```python
import torch
from torchvision import datasets, transforms
import torchvision.utils as vutils
from torch.distributions import Normal, kl_divergence, Bernoulli
from torch.utils.data import DataLoader, Dataset
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
from PIL import Image
import imageio
import os
import pdb
```

```python
class DspritesDataset(Dataset):
  def __init__(self, data_tensor):
    super(DspritesDataset, self).__init__()
    self.data_tensor = data_tensor
      
  def __len__(self):
    return self.data_tensor.shape[0]
  
  def __getitem__(self, idx):
    img = self.data_tensor[idx]
    return img
```

```python
root = os.path.join('./dataset', 'dsprites-dataset/dsprites_ndarray_co1sh3sc6or40x32y32_64x64.npz')
data = np.load(root, encoding='bytes')
imgs = torch.from_numpy(data['imgs']).unsqueeze(1).float()
idx = np.arange(imgs.shape[0])
np.random.shuffle(idx)
```

```python
num_train = int(imgs.shape[0] * 0.8)
train_data = imgs[idx[:num_train]]
test_data = imgs[idx[num_train:]]

train_ds = DspritesDataset(train_data)
test_ds = DspritesDataset(test_data)
```

```python
img_size = 64
z_dim = 10
class ChairDataset(Dataset):
  def __init__(self, root, split='train'):
    super(ChairDataset, self).__init__()
    num_train = 20000
    num_test = 4000
    img_files = sorted(glob(os.path.join(root, '*.png')))
    
    if split == 'train':
      self.img_files = img_files[: num_train]
    elif split == 'test':
      self.img_files = img_files[num_train : num_train + num_test]
    else:
      raise ValueError('Invalid split')
      
  def __len__(self):
    return len(self.img_files)
  
  def __getitem__(self, idx):
    with open(self.img_files[idx], 'rb') as f:
      with Image.open(f) as img:
        img = img.convert('RGB')
    
    transform = transforms.Compose([
      transforms.Grayscale(),
      transforms.Resize((img_size, img_size)),
      transforms.ToTensor()
    ])
    
    img = transform(img)
    return img
```

```python
class Encoder(nn.Module):
  def __init__(self, ):
    super(Encoder, self).__init__()
    self.conv_enc = nn.Sequential(
      nn.Conv2d(1, 32, 4, 2, padding=1),
      nn.ReLU(),
      nn.Conv2d(32, 32, 4, 2, padding=1),
      nn.ReLU(),
      nn.Conv2d(32, 64, 4, 2, padding=1),
      nn.ReLU(),
      nn.Conv2d(64, 64, 4, 2, padding=1),
      nn.ReLU(),
    )
    
    self.mlp_enc = nn.Sequential(
      nn.Linear(4*4*64, 256),
      nn.ReLU(),
      nn.Linear(256, 256),
      nn.ReLU(),
      nn.Linear(256, z_dim*2),
    )
    
  def forward(self, x, testing=True):
    conv_f = self.conv_enc(x)
    params = self.mlp_enc(conv_f.flatten(start_dim=1))
    
    mu, std = params.split(z_dim, dim=1)
    
    q = Normal(mu, F.softplus(std))
    
    p = Normal(mu.new_zeros(mu.shape), std.new_ones(std.shape))
    kl = kl_divergence(q, p)
    
    z = q.rsample()
    
    if testing:
      self.log_prob_p = p.log_prob(z)
      self.log_prob_q = q.log_prob(z)
    
    
    return z, kl
```

```python
class Decoder(nn.Module):
  def __init__(self, ):
    super(Decoder, self).__init__()
    self.mlp_dec = nn.Sequential(
      nn.Linear(z_dim, 256),
      nn.ReLU(),
      nn.Linear(256, 256),
      nn.ReLU(),
      nn.Linear(256, 4*4*64),
      nn.ReLU(),
    )
    self.conv_dec = nn.Sequential(
      nn.ConvTranspose2d(64, 64, 4, 2, padding=1),
      nn.ReLU(),
      nn.ConvTranspose2d(64, 32, 4, 2, padding=1),
      nn.ReLU(),
      nn.ConvTranspose2d(32, 32, 4, 2, padding=1),
      nn.ReLU(),
      nn.ConvTranspose2d(32, 1, 4, 2, padding=1),
    )
    
  def forward(self, z):
    mlp_f = self.mlp_dec(z)
    mu = self.conv_dec(mlp_f.reshape(-1, 64, 4, 4))
    
    return mu
```

```python
class VanillaVAE(nn.Module):
  def __init__(self, ):
    super(VanillaVAE, self).__init__()
    self.enc = Encoder()
    self.dec = Decoder()
    
  def forward(self, x, testing=False, k=50):
    z, kl = self.enc(x, testing=testing)
    mu = self.dec(z)
    
    rec_loss = self.comp_loss(x, mu)
    
    if testing:
      rec_p = Bernoulli(probs=mu.sigmoid())
#       rec_p = Normal(mu, mu.new_ones(mu.shape))
      ll = rec_p.log_prob(x).flatten(start_dim=1).sum(-1).view(-1, k)
      log_prob_p = self.enc.log_prob_p.sum(-1).view(-1, k)
      log_prob_q = self.enc.log_prob_q.sum(-1).view(-1, k)
      
      nll = -ll + log_prob_q - log_prob_p
      nll_max = nll.max(dim=-1, keepdim=True)[0]
      nll = (nll - nll_max).exp().mean(dim=-1, keepdim=True).log() + nll_max
      nll = nll.mean()
      return nll
    else:
      elbo = rec_loss.flatten(start_dim=1).sum(-1) + 4. * kl.flatten(start_dim=1).sum(-1)
      elbo = elbo.mean()
    
      return elbo, mu.sigmoid()
  
  def comp_loss(self, x, rec_mu):
    rec_loss = F.binary_cross_entropy_with_logits(rec_mu, x, reduction='none')
#     rec_loss = (x - rec_mu)**2
    return  rec_loss
```

```python
def eval_model(model, dataloader, k=50):
  nll = 0.
  model.eval()
  with torch.no_grad():
    for i, img in enumerate(dataloader):
      B, C, H, W = img.shape
      img = img.unsqueeze(1).expand(-1, k, -1, -1, -1).reshape(B*k, C, H, W)
      img = img.to(device)
      nll_batch = model(img, testing=True)
      nll += nll_batch.mean().detach()
    
  nll = nll / float(len(dataloader))
  return nll
```

```python
epochs = 100
nll_list = []
train_loss = []
for e in range(epochs):
  net, loss = train_model(net, dl_train, optimizer, e)
  train_loss.append(loss.cpu().item())

  nll = eval_model(net, dl_test)
  nll_list.append(nll.cpu().item())
  print('Epoch:\t{}/{},\tnll:{:.6f}'.format(e, epochs, np.mean(nll_list)))
  if (e+1) % 10 == 0:
    plt.plot(np.arange(e+1), train_loss, 'r-', np.arange(e+1), nll_list, 'g-')
    plt.show()
```

```python
def traversal(net, img):
  z, _ = net.enc(img)
  
  interpolation = torch.linspace(-2., 2., steps=10, device=z.device)
  print(interpolation.shape)
  rec_mus = []
  for dim in range(z_dim):
    for val in interpolation:
      z[:, dim] = val
      mu = net.dec(z)
      rec_mus.append(mu.sigmoid().detach())
    
  return rec_mus
```
